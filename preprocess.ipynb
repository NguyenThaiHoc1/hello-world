```python
import os #get path
import numpy as np
import glob
import cv2
import progressbar
import pickle
from align_dlib import AlignDlib
```

```python
number_of_image = 1
root_current_folder = os.path.abspath('')
folder_dataset = os.path.join(root_current_folder, 'dataset', 'lfw')
dlib_landmark_model = os.path.join(root_current_folder, 'dlib_landmark_model')
extensions_vali = {".jpg", ".png", '.jpeg'}
align_dlib = AlignDlib(os.path.join(dlib_landmark_model, 'model_landmark'))

print("root_folder is: '{}'".format(root_current_folder))
print("dataset folder is: '{}'".format(folder_dataset))
```

```python
count = 0
output_dir_preprocess = os.path.join(root_current_folder, 'dataset', 'lfw_preprocess')
for image_dir in os.listdir(folder_dataset):
    count += 1
    image_output_dir = os.path.join(output_dir_preprocess, os.path.basename(os.path.basename(image_dir)))
    if not os.path.exists(image_output_dir):
        os.makedirs(image_output_dir)
        
print("-- Create Done --- {}".format(count))
```

```python
# ham doc toan folder cua mot dataset
def read_all_folders(path_folder):
    return [compine for compine in os.walk(path_folder)]

# check phan mo rong cua file co hop le hay khong
def check_file_vali(name_of_file):
    status_checked = any(name_of_file.endswith(ext) for ext in extensions_vali)
    return status_checked

def read_image(path_filename):
    image = cv2.imread(path_filename, 1)
    if image is not None:
        image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    return image
    
    
def preprocess(np_image, crop_dim=160):
    bb = align_dlib.getLargestFaceBoundingBox(np_image) # khuc nay la khuc detection na 
    aligned = align_dlib.align(crop_dim, np_image, bb, landmarkIndices=AlignDlib.INNER_EYES_AND_BOTTOM_LIP)
    if aligned is not None:
        aligned = cv2.cvtColor(aligned, cv2.COLOR_BGR2RGB)
    return aligned
    
    
def write_image(folder_name, file_name, np_image):
    path_name_file = os.path.join(output_dir_preprocess, folder_name, file_name)
    cv2.imwrite(path_name_file, np_image)

def read_image_from_folder(name_each_folder):
    class_image = []
    os_path_name = os.path.join(folder_dataset, str(name_each_folder))
    print("+ Name: {}".format(name_each_folder))
    # hoac o day co the su dung os.listdir --- tips: glob.glob: abs_path chung ta muon lay file name thi os.path.basename(file)
    for file in glob.glob(os.path.join(os_path_name, '*')):
        file_name = os.path.basename(file)
        if not check_file_vali(file_name):
            continue
        try:
            image = read_image(file)
        except:
            continue
        if image is None:
            continue
        image = preprocess(image)
        write_image(name_each_folder, file_name, image)
        class_image.append(image)
        print("  - {}".format(file))
    return class_image
        
        
```

```python
images = []
labels = []
names = []
count = 0
list_folders_name = read_all_folders(folder_dataset)[0][1]
for folder in list_folders_name:
    class_image = read_image_from_folder(folder)
    if len(class_image) > number_of_image:
        names.append(folder)
        for img in class_image:
            images.append(img)
            labels.append(count)
        count += 1
    else:
        print("  * The number of image is not enough -- name: {}".format(folder))
```

```python
path_folder_name_pickle = os.path.join(root_current_folder, "data_preprocess_pickle")
folder_name = [os.path.join(path_folder_name_pickle, folder_name) for folder_name in ['pickle_train.pkl', 'pickle_labels.pkl', 'pickle_names.pkl']]
with open(folder_name[0], 'wb') as handle:
    pickle.dump(images, handle, protocol=pickle.HIGHEST_PROTOCOL)
    
with open(folder_name[1], 'wb') as handle:
    pickle.dump(labels, handle, protocol=pickle.HIGHEST_PROTOCOL)
    
with open(folder_name[2], 'wb') as handle:
    pickle.dump(names, handle, protocol=pickle.HIGHEST_PROTOCOL)
    
```
